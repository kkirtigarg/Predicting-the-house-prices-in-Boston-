```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

### When data is not scaled.

```python
def step_gradient(points, learning_rate, m):
    mslope = np.zeros(14)
    M = len(points)
    for i in range(M):
        x = np.array(points[i, 0:14])
        y = points[i,14]
        mslope += (-2/M)*(y-(m*x).sum())*x
    new_m = m - learning_rate*mslope
    return new_m
        
    
```

```python
def cost(points, m):
    C = 0
    M = len(points)
    for i in range(M):
        x = np.array(points[i,0:14])
        y = points[i,14]
        C += (1/M)*((y-(m*x).sum())**2)
    return C
```

```python
def gd(points, learning_rate, num_iterations):
    m = np.zeros(14)
    for i in range(num_iterations):
        m = step_gradient(points, learning_rate, m)
#         print(i, "cost : ", cost(points,m))
    return m
        
```

```python
# when data is not scaled.
def run():
    train = np.genfromtxt("boston_train.csv", delimiter = ",")
    df = pd.DataFrame(train)
    df.insert(loc = 13, column = "14", value = [1 for i in range(379)])
    train_ = df.values
    learning_rate = 0.1
    num_iterations = 800
    m = gd(train_, learning_rate, num_iterations)
    print(m)
    return m
    
    
```

```python
def predict(m):
    x_test = np.genfromtxt("boston_test.csv", delimiter = ",") 
    df = pd.DataFrame(x_test)
    df['13'] = 1
    x = df.values
    y_pred = []
    for i in range(len(x_test)):
        y_pred.append((m*x[i,:]).sum())
    print(y_pred) 
    return y_pred
    
```

### When data is scaled.

```python
def step_gradient(x_train, y_train, learning_rate, m):
    mslope = np.zeros(14)
    M = len(x_train)
    for i in range(M):
        x = np.array(x_train[i,:])
        y = y_train[i]
        mslope += (-2/M)*(y-(m*x).sum())*x
    new_m = m - learning_rate*mslope
    return new_m
        
    
```

```python
def cost(x_train, y_train, m):
    C = 0
    M = len(x_train)
    for i in range(M):
        x = np.array(x_train[i,:])
        y = y_train[i]
        C += (1/M)*((y-(m*x).sum())**2)
    return C
```

```python
def gd(x_train, y_train, learning_rate, num_iterations):
    m = np.zeros(14)
    for i in range(num_iterations):
        m = step_gradient(x_train, y_train, learning_rate, m)
#         print(i, "cost : ", cost(x_train, y_train,m))
    return m
        
```

```python
# when data is scaled.
def run_scaled_data():
    train = np.genfromtxt("boston_train.csv", delimiter = ",")
    scaler = preprocessing.StandardScaler()
    scaler.fit(train[:,0:13])
    train_scaled = scaler.transform(train[:,0:13])
    df = pd.DataFrame(train_scaled)
    df['14'] = 1
    xtrain = df.values
    learning_rate = 0.1
    num_iterations = 700
    m = gd(xtrain, train[:,13], learning_rate, num_iterations)
    print(m)
    return m
```

```python
def predict(m):
    x_test = np.genfromtxt("boston_test.csv", delimiter = ",") 
    df = pd.DataFrame(x_test)
    df['13'] = 1
    x = df.values
    y_pred = []
    for i in range(len(x_test)):
        y_pred.append((m*x[i,:]).sum())
    print(y_pred) 
    return y_pred
```

```python
m2 = run_scaled_data()
```

```python
y_pred_scaled = predict(m2)
print(y_pred_scaled)
```
